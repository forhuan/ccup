```python
# load the image and convert it to grayscale
image = cv2.imread('./imgs/i030.png')
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray, cmap=plt.get_cmap('gray'), vmin=0, vmax=255)
```

```python
blurred = cv2.GaussianBlur(gray, (65, 65), 0)
(minVal, maxVal, minLoc, maxLoc) = cv2.minMaxLoc(blurred)
circled = np.copy(image)
cv2.circle(circled, maxLoc, 64, (255, 0, 0), 5)
```

```python
lp = (0,0)
misses = 0
for n in range(1,1000):
    fname = "i{0:0=3d}.png".format(n)
    try:        
        image = cv2.imread('./imgs/{}'.format(fname))
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
        blurred = cv2.GaussianBlur(gray, (65, 65), 0)
        (minVal, maxVal, minLoc, maxLoc) = cv2.minMaxLoc(blurred)
        circled = np.copy(image)
        if (maxVal > 8.0):
            cv2.circle(circled, maxLoc, 64, (255, 0, 0), 5)
            lp = maxLoc
            misses = 0
        elif misses < 5:
            cv2.circle(circled, lp, 64, (255, 0, 0), 5)
            misses += 1
        cv2.imwrite('out/{}'.format(fname), circled)
    except:
        print "Done with", fname
        break;
```
